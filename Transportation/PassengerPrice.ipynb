```python
import json
import datetime
with open('transportation.usa_airline_fuel_cost_consumption.json', 'r') as f:
     allDataPrice = json.load(f)

#print allData
data = allDataPrice['results'][0]
rawDataPrice = data['data']

monthPrice = []

for item in rawDataPrice:
    dateTime = datetime.datetime.fromtimestamp(item[0]/10000).strftime('%Y-%m-%d %H:%M:%S')
    splitData = dateTime.split()
    monthYear = splitData[0]
    monthPrice.append([monthYear, item[9]])

sorted(monthPrice, key=itemgetter(0), reverse=True)
#print monthPrice
```

```python
from operator import itemgetter, attrgetter, methodcaller

with open('transportation.monthy_number_of_airline_passengers_all_carriers.json', 'r') as f:
     allDataPassenger = json.load(f)
        
data = allDataPassenger['results'][0]
rawDataPassenger = data['data']

monthPassenger = []

for item in rawDataPassenger:
    dateTime = datetime.datetime.fromtimestamp(item[0]/10000).strftime('%Y-%m-%d %H:%M:%S')
    splitData = dateTime.split()
    monthYear = splitData[0]
    monthPassenger.append([monthYear,item[3]])

sorted(monthPassenger, key=itemgetter(0), reverse=True)
```
